```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Healthcare_dataset.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
for i in df.select_dtypes(include='object').columns:
    if df[i].nunique()<=20:
        print("Value counts for {}:".format(i),df[i].value_counts())
        print("--"*50)
        df[i].value_counts().plot(kind='bar')
        plt.show()
```

```python
sns.histplot(data=df,x='Billing Amount',kde=True)
```

```python
sns.histplot(data=df,x='Age',kde=True)
```

```python
sns.countplot(data=df,x='Blood Type',hue='Gender')
```

```python
fig,ax=plt.subplots(1,4,figsize=(15,8))
ax[0].pie(df['Medical Condition'].value_counts(),labels=df['Medical Condition'].unique(),autopct='%1.1f%%')
ax[0].set_title('Medical COndition')

ax[1].pie(df['Insurance Provider'].value_counts(),labels=df['Insurance Provider'].unique(),autopct='%1.1f%%')
ax[1].set_title('Insurance Provider')

ax[2].pie(df['Admission Type'].value_counts(),labels=df['Admission Type'].unique(),autopct='%1.1f%%')
ax[2].set_title('Admission Type')

ax[3].pie(df['Test Results'].value_counts(),labels=df['Test Results'].unique(),autopct='%1.1f%%')
ax[3].set_title('Test Results')

plt.tight_layout()
plt.show()
```
